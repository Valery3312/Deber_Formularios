##LOC
- Filtra por etiquetas
###.loc[] se usa para seleccionar filas y columnas por el nombre del índice o de la columna.
- df.loc[fila, columna]


- Características:
1. Usa etiquetas (nombres de filas y columnas).
2. Incluye el valor final del rango (es inclusivo).
3. Permite condiciones lógicas (boolean indexing).

```python
# ["Timestamp","Hours per day","BPM","Permissions"]] → indica que sólo queremos la información de estas columnas
#[0:4,....] →indica que se seleccionan las filas desde la 0 hasta la 4, incluyendo la 4 (a diferencia de iloc o NumPy que excluyen el último valor).
df_music.loc[0:4, ["Timestamp","Hours per day","BPM","Permissions"]]
```

##Indexado básico (slice)
- Filtra por posición de las filas

```python
#[0:4] → indica que se seleccionan las filas desde la posición 0 hasta la 3 (la fila 4 no se incluye).
#El resultado es un nuevo DataFrame con las primeras 4 filas (índices 0, 1, 2 y 3) y todas las columnas.
df_music[0:4]
```

##ILOC
- Filtrado por posición numérica
-.iloc[] se usa para seleccionar filas y columnas por posición entera (índices numéricos).

-Características:
1. Usa Índices enteros (posición)
2. No incluye el valor final del rango (es exclusivo).
3. No permite condiciones lógicas directamente.

```python
df_music.iloc[0:4, [0,1,2,3]] #No se muestra el 4 (enrtonces el 3 es 4)
```

## Vamos a rellenar valores y limpiaremos si es necesario
- Usaremos la mediana, no la media porque es muy alta.

El máximo de pulsaciones que puede relizar un género musical es de 200a 300 MPB
- Por eso verificaremos si hay outilers mayores a 300

```python
#Remplazamos nulos por la mediana
df_music["BPM"] = df_music["BPM"].fillna(df_music["BPM"].median())
```

```python
# Verificiamos cuantos nulos quedan → 22 de 129
total_nulos = df_music.isnull().sum().sum()
print("\nTotal de valores nulos en la base:", total_nulos)
```

```python
# Revisamos cuantos outilers > 300 tenemos
valores_mayores_300 = (df_music["BPM"] > 300).sum()
# Mostrar el resultado
print(f"Cantidad de valores BPM > 300: {valores_mayores_300}")
# No tenemos outilers > 300
```
